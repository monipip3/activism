```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd
from tabulate import tabulate
import html5lib
```

```python
url = 'https://www.dshs.texas.gov/CHS/VSTAT/VS07/t14a.shtm'
page = requests.get(url,verify=False)

page_text = page.text

b = BeautifulSoup(page_text,'lxml')
```

```python
table = b.find_all('table')
```

```python
df = pd.read_html(str(table),header=0)[0]
```

```python
df = df.convert_objects(convert_numeric=True)
```

```python
df = df.iloc[:,0:2]
```

```python
df['Year'] = '2007'
```

```python
urls = ['https://www.dshs.texas.gov/CHS/VSTAT/VS07/t14a.shtm','https://www.dshs.state.tx.us/chs/vstat/latest/t14a.shtm','https://www.dshs.texas.gov/chs/vstat/vs09/t14A.shtm'
       ,'https://www.dshs.state.tx.us/chs/vstat/vs10/t14a.shtm','https://www.dshs.state.tx.us/chs/vstat/vs11/t14a.shtm','https://www.dshs.state.tx.us/chs/vstat/vs12/t14a.shtm',
        'https://www.dshs.state.tx.us/chs/vstat/vs13/t14a.aspx', 'https://www.dshs.state.tx.us/chs/vstat/vs14/t14a.aspx',
        'https://www.dshs.state.tx.us/chs/vstat/vs15/t14a.aspx'
       ]
```

```python
def make_tables(x,y,df):
    page = requests.get(x,verify=False)
    page_text = page.text
    b = BeautifulSoup(page_text,'lxml')
    table = b.find_all('table')
    df2 = pd.read_html(str(table),header=0)[0]
    df2 = df2.convert_objects(convert_numeric=True)
    df2 = df2.iloc[:,0:2]
    df2['Year'] = 2007 + y
    df = df.append(df2)
    return df
```

```python
df = make_tables(urls[2],2,df)
df = make_tables(urls[3],3,df)
df = make_tables(urls[4],4,df)
df = make_tables(urls[5],5,df)
df = make_tables(urls[6],6,df)
df = make_tables(urls[7],7,df)
```

```python
df = make_tables(urls[8],8,df)
```
